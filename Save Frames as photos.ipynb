```python
# compress video from avi to mp4 (h.264)
# not as fast as matlab

totTm = time.time()

cap = cv2.VideoCapture(aa)

print 'numFrames = ' + str(cap.get(cv2.cv.CV_CAP_PROP_FRAME_COUNT))


# re-set frame number to 0
cap.set(cv2.cv.CV_CAP_PROP_FRAME_COUNT, 0) 

w=int(cap.get(cv2.cv.CV_CAP_PROP_FRAME_WIDTH ))
h=int(cap.get(cv2.cv.CV_CAP_PROP_FRAME_HEIGHT ))
fps = cap.get(cv2.cv.CV_CAP_PROP_FPS)

# Define the codec and create VideoWriter object

# video recorder
#fourcc = cv2.cv.CV_FOURCC(*'mp4v') # slow
fourcc = cv2.cv.CV_FOURCC(*'avc1') # 3 x slower than divx -- save as .mp4
#fourcc = cv2.cv.CV_FOURCC(*'divx') # fast, doesn't work, though
#fourcc = cv2.cv.CV_FOURCC(*'xvid') # fast, but doesn't actually save a video

out = cv2.VideoWriter("~/Desktop/testOutput5.mp4", fourcc, fps, (w, h))
#out = cv2.VideoWriter('output', -1, 30, (int(height), int(width)), isColor = True)

#out = cv2.VideoWriter('output1', fourcc, 30, (int(height), int(width)), isColor = False)

ctr = 1

while(cap.isOpened()):
    strr = time.time()
    ret, frame = cap.read()
    
    if ret==True:
        #out.write(frame)
        #convert to grayscale
        gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
        out.write(np.dstack([gray,gray,gray])) #have to write 3 channels
        stopR = time.time() - strr
        print("counter = " + str(ctr) + " time = " + str(stopR))

        
        ctr += 1
    else:
        break

print str( time.time() - totTm)
# Release everything if job is finished
cap.release()
out.release()


```

```python
# adjust gamma
def adjust_gamma(image, gamma=1.0):
    # build a lookup table mapping the pixel values [0, 255] to
    # their adjusted gamma values
    invGamma = 1.0 / gamma
    table = np.array([((i / 255.0) ** invGamma) * 255
        for i in np.arange(0, 256)]).astype("uint8")
 
    # apply gamma correction using the lookup table
    return cv2.LUT(image, table)
```
